```python
class Patient:
    def __init__(self, id, name, age, blood_group):
        self.id=id
        self.name=name
        self.age=age
        self.blood_group=blood_group
        self.prev=None
        self.next=None


class WRM:
    def __init__(self):
        self.head = Patient(None,None,None,None)
        self.head.prev=self.head
        self.head.next=self.head
        self.size=0

    def registerPatient(self,id,name,age,blood_group):
        new_patient=Patient(id,name,age,blood_group)
        new_patient.prev=self.head
        new_patient.next=self.head.next
        self.head.next.prev=new_patient
        self.head.next=new_patient
        self.size+=1
        print('Success')

    def servePatient(self):
        if self.size == 0:
            print('No patients to serve.')
            return
        patient_to_serve=self.head.next
        self.head.next=patient_to_serve.next
        patient_to_serve.next.prev=self.head
        self.size-=1
        print(f"Serving Patient: {patient_to_serve.name}")

    def cancelAll(self):
        self.head.next=self.head
        self.head.prev=self.head
        self.size=0
        print('Success')

    def canDoctorGoHome(self):
        return self.size==0

    def showAllPatients(self):
        if self.size==0:
            print('No patients waiting.')
            return
        current = self.head.next
        while current != self.head:
            print(current.id, end=" ")
            current=current.next
        print()

    def reverseTheLine(self):
        if self.size==0:
            print('No patients to reverse.')
            return
        current=self.head.prev
        while current!=self.head:
            print(current.id, end=" ")
            current=current.prev
        print()
        print('Success')

def main():
    wrm = WRM()
    while True:
        print("\nOptions:")
        print("1. Register Patient")
        print("2. Serve Patient")
        print("3. Show All Patients")
        print("4. Can Doctor Go Home?")
        print("5. Cancel All Patients")
        print("6. Reverse The Line")
        print("7. Exit")
        choice = int(input("Enter your choice: "))

        if choice == 1:
            id = int(input("Enter Patient ID: "))
            name = input("Enter Patient Name: ")
            age = int(input("Enter Patient Age: "))
            blood_group = input("Enter Patient Blood Group: ")
            wrm.registerPatient(id, name, age, blood_group)

        elif choice == 2:
            wrm.servePatient()

        elif choice == 3:
            wrm.showAllPatients()

        elif choice == 4:
            if wrm.canDoctorGoHome():
                print("Yes")
            else:
                print("No")

        elif choice == 5:
            wrm.cancelAll()

        elif choice == 6:
            wrm.reverseTheLine()

        elif choice == 7:
            break

        else:
            print("Invalid choice. Please try again.")


if __name__ == "__main__":
    main()
```

```python
class Node:
    def __init__(self,letter):
        self.letter=letter
        self.next=None

class LinkedList:
    def __init__(self):
        self.head=None
    def append(self,letter):
        nNode=Node(letter)
        if not self.head:
            self.head=nNode
            return
        last=self.head
        while last.next:
            last=last.next
        last.next=nNode

    def length1(self):
        count=0
        current=self.head
        while current:
            count+=1
            current=current.next
        return count

    def decoded_word(self):
        length=self.length1()
        if length==0:
            return None
        modValue=13 % length
        positions=[i for i in range(length) if i % modValue==0]
        decodedLetters=[]
        for pos in positions:
            current=self.head
            for _ in range(pos):
                if current:
                    current=current.next
            if current:
                decodedLetters.append(current.letter)

        decodedLetters.reverse()
        result=LinkedList()
        result.append("None")
        for letter in decodedLetters:
            result.append(letter)

        return result

    def display(self):
        current = self.head
        result = []
        while current:
            result.append(current.letter)
            current = current.next
        return '->'.join(result) if result else 'None'

#B->M->D->T->N->O->A->P->S->C
if __name__ == "__main__":
    linked_list = LinkedList()
    input_letters = input("Enter the letters:").strip().split('->')
    for letter in input_letters:
        linked_list.append(letter)

    decoded_list = linked_list.decoded_word()

    print("Resulting Linked List:", decoded_list.display())
```

```python
#TASK3
class ListNode:
    def __init__(self,val=0,next=None):
        self.val=val
        self.next=next

def reverse_linked_list(head):
    prev = None
    while head:
        next_node = head.next
        head.next = prev
        prev = head
        head = next_node
    return prev

def add_Llists(h1, h2):
    dummy=ListNode(0)
    current=dummy
    while head1 and h2:
        sum_val=(h1.val+h2.val) % 10
        current.next=ListNode(sum_val)
        current=current.next
        h1=h1.next
        h2=h2.next
    return dummy.next

def idGenerator(h1,h2,h3):
    h1=reverse_linked_list(h1)
    sum_head=add_Llists(h2,h3)
    dummy=ListNode(0)
    current=dummy
    while h1:
        current.next=ListNode(h1.val)
        current =current.next
        h1=h1.next
    while sum_head:
        current.next = ListNode(sum_head.val)
        current =current.next
        sum_head =sum_head.next
    return dummy.next
def cLinked_list(lst):
    dummy=ListNode(0)
    current=dummy
    for num in lst:
        current.next=ListNode(num)
        current=current.next
    return dummy.next
def linkedList(head):
    result = []
    while head:
        result.append(str(head.val))
        head = head.next
    print("->".join(result))

# inputs
h1 =cLinked_list([0, 3, 2, 2])
h2= cLinked_list([5, 2, 2, 1])
h3 =cLinked_list([4, 3, 2, 1])
result_head=idGenerator(h1,h2,h3)
linkedList(result_head)
```
